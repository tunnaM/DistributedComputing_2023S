```python
url1 = "https://www.bilibili.com/video/BV1FM4y1k7Nk/?spm_id_from=333.1007.tianma.1-2-2.click&vd_source=8a00dab0be94d29388f2286892ba8d50"
```

```python
def get_bv_from_url(url):
    bv = url.split("/")
    bv = list(filter(lambda s: s.startswith("BV"),bv))
    if len(bv) < 1:
        raise RuntimeError("BV warning")
    bv = bv[0]
    qmark = bv.find("?") # 确定是否含“？”
    if qmark >= 0:
        bv = bv[:qmark] # 取索引，此处表示取得qmark之前的内容
    return bv
```

```python
bv = get_bv_from_url(url1)
bv
```

```python
api = "https://api.bilibili.com/x/web-interface/view?bvid=" 
url = api + bv 
print(url) # 该url包含弹幕相关信息
```

```python
#固定code读取网页内容
import urllib.request
response = urllib.request.urlopen(url)
string = response.read()
html = string.decode("utf-8") #返回的数据类型为json
print(html)
```

```python
import json
doc = json.loads(html) # 返回的是python中的字典，对json格式内容进行解析
doc # 通过观察，可在字典中确定cid的位置
cid = doc["data"]['cid']
print(cid) #得到了当前网站的cid
```

```python
api = "https://api.bilibili.com/x/v1/dm/list.so?oid="
url = api + str(cid)
url
```

```python
response = urllib.request.urlopen(url)
html = response.read() #返回的数据类型为json
print(html) # 这里是加密的
```
