```python
bv = list(filter(lambda s: s.startswith("BV"),bv))
bv
```

```python
def get_bv_from_url(url):
    bv = url.split("/") # 列表，其中包含字符串
    bv = list(filter(lambda s: s.startswith("BV"),bv)) # 找到以BV开始的字符串
    if len(bv) < 1:
        raise RuntimeError("BV warning") # 这个特别注意一下，自己生成警告
    bv = bv[0]
    qmark = bv.find("?") # 确定是否含“？”
    if qmark >= 0:
        bv = bv[:qmark] # 取索引，此处表示取得qmark之前的内容
    return bv
```

```python
bv = get_bv_from_url(url1)
bv
```

```python
import urllib.request
import json
def get_cid_from_bv(bv):
    api = "https://api.bilibili.com/x/web-interface/view?bvid="
    url = api + bv
    response = urllib.request.urlopen(url)
    string = response.read()
    html = string.decode("utf-8") #返回的数据类型为json
    doc = json.loads(html)
    cid = doc["data"]['cid']
    return cid
```

```python
cid = get_cid_from_bv(bv)
cid
```

```python
import zlib
import xml.dom.minidom

def get_xml_from_cid(cid):
    url = api + str(cid)
    response = urllib.request.urlopen(url)
    response.getheader("Content-Encoding")
    string = response.read()
    html = zlib.decompress(string, wbits=-zlib.MAX_WBITS).decode ('utf-8')
    xml2 = xml.dom.minidom.parseString(html)
    xml_pretty_str = xml2.toprettyxml()
    return xml_pretty_str
```
